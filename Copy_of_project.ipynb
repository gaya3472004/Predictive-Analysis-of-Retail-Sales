```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def subtotal(self):
        return self.price * self.quantity

class Superstore:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def calculate_total(self):
        total = 0
        for product in self.products:
            total += product.subtotal()
        return total

# Example usage
superstore = Superstore()

product1 = Product("Chair", 50, 2)
product2 = Product("Table", 100, 1)

superstore.add_product(product1)
superstore.add_product(product2)

print("Superstore Inventory:")
for product in superstore.products:
    print(f"{product.name}: ${product.price} x {product.quantity} = ${product.subtotal()}")

print(f"Total: ${superstore.calculate_total()}")
```
